```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
from matplotlib import rcParams
rcParams.update({'font.size': 18})
plt.rcParams['figure.figsize'] = [8,8]
```

```python
fig,ax = plt.subplots(1)
x = np.arange(0.2,5,0.1) #ranging from 0.2 to 4.9 with an increment of 0.1, number of terms in the model
y = np.divide(1,x) #pareto frontier curve
n = len(x) #48
y2 = y + 0.5*np.random.randn(n) #48 good models, close to the pf curve

y3 = (np.tile(y2,(5,1)) + 2*np.random.rand(5,n) + 1) #48times5 bad models
y3 = np.reshape(y3,-1)
x3 = np.tile(x,(1,5)).reshape(-1) #for plotting

plt.plot(x,y,color='k',linewidth=2)

rect = Rectangle((0.5,0.4), 0.9, 1.6,linewidth=1,edgecolor='k',facecolor='grey',alpha=0.6) #pareto optimal region
ax.add_patch(rect)

plt.scatter(x,y2,100,color='lime',edgecolors='k')
plt.scatter(x3,y3,100,color='blue',edgecolors='k')


plt.xlim(0.2,4)
plt.ylim(0,5.5)
plt.xlabel('Number of Terms (Complexity)',fontweight='bold', fontsize=20)
plt.ylabel('Error Measurement',fontweight='bold', fontsize=20)


plt.text(2.5, 4.25, 'Models', color='Red',fontsize=17)
plt.text(2.5, 0.5, 'Pareto Frontier', color='Red',fontsize=17)
plt.text(0.55, 0.5, 'Pareto Optimal', color='Red', fontsize=13)
#plt.tight_layout()
plt.show()
```
